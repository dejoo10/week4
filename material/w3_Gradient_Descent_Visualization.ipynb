## 2D Gradient Descent Visualization
In a 2D plot, you can visualize the loss surface and the steps taken by gradient descent to find the minimum.

````python
import numpy as np
import matplotlib.pyplot as plt

# Define the loss function (example: quadratic function)
def loss_function(x):
    return x**2

# Gradient descent function
def gradient_descent_2d(learning_rate, epochs):
    x = 10  # Initial guess

    # Lists to store x and loss values for visualization
    x_values = [x]
    loss_values = [loss_function(x)]

    # Gradient descent loop
    for epoch in range(epochs):
        gradient = 2 * x  # Gradient of the loss function
        x -= learning_rate * gradient  # Update rule
        x_values.append(x)
        loss_values.append(loss_function(x))

    return x_values, loss_values

# Parameters
learning_rate = 0.1
epochs = 20

# Run gradient descent
x_values, loss_values = gradient_descent_2d(learning_rate, epochs)

# Plot
plt.figure(figsize=(10, 6))
plt.plot(x_values, loss_values, '-o', color='b', markerfacecolor='r', label='Gradient Descent Path') # Line in blue, circles in red

plt.title('2D Gradient Descent Visualization')
plt.xlabel('Parameter x')
plt.ylabel('Loss')
plt.grid(True)
plt.legend()  # Add legend to display the label
plt.show()
````

## 3D Gradient Descent Visualization
In a 3D plot, you can visualize the loss surface and the steps taken by gradient descent to find the minimum.

````python
from mpl_toolkits.mplot3d import Axes3D

# Define the loss function (example: quadratic function in 3D)
def loss_function_3d(x, y):
    return x**2 + y**2

# Gradient descent function in 3D
def gradient_descent_3d(learning_rate, epochs):
    x, y = 10, 10  # Initial guesses

    # Lists to store x, y, and loss values for visualization
    x_values, y_values, loss_values = [x], [y], [loss_function_3d(x, y)]

    # Gradient descent loop
    for epoch in range(epochs):
        gradient_x = 2 * x  # Gradient of the loss function w.r.t. x
        gradient_y = 2 * y  # Gradient of the loss function w.r.t. y
        x -= learning_rate * gradient_x  # Update rule for x
        y -= learning_rate * gradient_y  # Update rule for y
        x_values.append(x)
        y_values.append(y)
        loss_values.append(loss_function_3d(x, y))

    return x_values, y_values, loss_values

# Parameters
learning_rate = 0.1
epochs = 20

# Run gradient descent
x_values, y_values, loss_values = gradient_descent_3d(learning_rate, epochs)

# Create meshgrid for 3D plot
X, Y = np.meshgrid(np.linspace(-12, 12, 100), np.linspace(-12, 12, 100))
Z = loss_function_3d(X, Y)

# Plot
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.6)
ax.scatter(x_values, y_values, loss_values, color='r', s=50, label='Gradient Descent Path')
ax.set_title('3D Gradient Descent Visualization')
ax.set_xlabel('Parameter x')
ax.set_ylabel('Parameter y')
ax.set_zlabel('Loss')
ax.legend()
plt.show()
````
